```python
import cv2 as cv
import os
video=cv.VideoCapture(0)
face=cv.CascadeClassifier('haarcascade_frontalface_default.xml')
z=0
path='images/'
isExist = os.path.exists(path)
os.makedirs(path)

while True:
	ret,frame=video.read()
	faces=face.detectMultiScale(frame,1.3, 5)
	for x,y,w,h in faces:
		z=z+1
		name='./images/'+ str(z) + '.jpg'
		print("Creating Images........." +name)
		cv.imwrite(name, frame[y:y+h,x:x+w])
		cv.rectangle(frame, (x,y), (x+w, y+h), (0,255,255), 5)
	cv.imshow("FRAME", frame)
	cv.waitKey(2)
	if z>4:
		break
video.release()
cv.destroyAllWindows()

from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np

new_model = load_model('models/genderclassification.h5')
sum=0

for i in range (1,6):
    
    img = cv.imread('images/'+str(i)+'.jpg')
    resize = tf.image.resize(img, (256,256))
    
  
    sum =sum+  new_model.predict(np.expand_dims(resize/255, 0))
    
 
yhat=sum/5
if yhat >= 0.5: 
    gender=0
else:
    gender=1

path = 'images'
import shutil
shutil.rmtree(path)

font = cv.FONT_HERSHEY_SIMPLEX
scale = 1
color = (0, 0,255)
thickness = 5

def human(gender):
    if gender==0:
        return ('Male')
    else:
        return ('Female')

cap = cv.VideoCapture(0)

while True:
   
    ret, frame = cap.read()
    cv.rectangle(frame, (x,y), (x+w, y+h), (0,255,255), 5)
    cv.putText(frame, human(gender), (50, 50), font, scale, color, thickness)
    cv.imshow('frame', frame)

    if cv.waitKey(1) == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
